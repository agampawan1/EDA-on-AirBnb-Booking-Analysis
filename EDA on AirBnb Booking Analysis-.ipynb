# Project Summary -

Airbnb (ABNB) is an online marketplace that connects people who want to rent out their homes with people who are looking for accommodations in specific locales.Airbnb offers people an easy, relatively stress-free way to earn some income from their property

This project is an exploratory data analysis (EDA) on Airbnb booking data. The goal of the analysis is to gain insights into the patterns and trends of Airbnb bookings in the city of New York. The dataset used for the analysis contains information on the listings, reviews, and bookings of Airbnb properties in New York City from 2019.

**4.Which Hosts are charging higher price?**

```python
higher_price=df.groupby(['host_id','host_name','room_type','neighbourhood_group'])['price'].max().reset_index()
higher_price=higher_price.sort_values(by='price',ascending=False).head(12)
higher_price
```

**8.What is the average nightly price of Airbnb listings in New York City, and how does it vary by borough**

```python
average_nightly_price=df.groupby('neighbourhood_group')['price'].mean()
average_nightly_price
```

**Visualizing the above results using a bar chart.**

```python
plt.bar(average_nightly_price.index, average_nightly_price.values)
plt.title('Average Nightly Price by Borough')
plt.xlabel('Borough')
plt.ylabel('Average Nightly Price')
plt.show()
```

**From above chart we can see that Manhattan Borough has highest Average Nightly Price of $196.875814**

**9.Now analysing 'Price' variable with 'latitude' variable and 'longitude' variable of the data**.

```python
#import librarie
import folium

# Create a map centered on New York City
map_nyc = folium.Map(location=[40.7128, -74.0060], zoom_start=11)

# Add markers for each listing, colored by price
for lat, lon, price in zip(df['latitude'], df['longitude'], df['price']):
    color = 'green'
    if price > 200:
        color = 'red'
    marker = folium.CircleMarker(location=[lat, lon], radius=5, color=color, fill=True, fill_color=color)
    marker.add_to(map_nyc)

# Display the map
map_nyc
```
